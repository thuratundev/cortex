 **"Embedding Model ကို အဓိပ္ပာယ်အပြည့်စုံဆုံးနဲ့ အရှင်းလင်းဆုံးဖြစ်မယ့် စာသားတစ်တုံး (a single block of text) ဖြစ်အောင် ပြင်ဆင်ပြီးမှ ပေးပို့ပါ"** ဆိုတာပါပဲ။

Model က JSON object တစ်ခုလုံးကို တိုက်ရိုက်နားလည်တာမဟုတ်ဘဲ၊ သင်က အဲဒီ JSON ထဲက အချက်အလက်တွေကို ပေါင်းစပ်ပြီး ပေးလိုက်တဲ့ **စာသား (text)** ကိုသာ နားလည်ပြီး Vector ပြောင်းပေးတာပါ။

---

### အထိရောက်ဆုံး JSON Format နှင့် ပြင်ဆင်နည်းများ

သင့်ရဲ့ JSON object တစ်ခုချင်းစီကို Embedding မလုပ်ခင် အောက်ပါပုံစံအတိုင်း စာသားတစ်တုံးအဖြစ် ပြောင်းလဲပါ။

#### နည်းလမ်း ၁: Label တပ်ပြီး အချက်အလက်အားလုံးကို ပေါင်းစပ်ခြင်း (The Best Method)

ဒါက အထိရောက်ဆုံးနည်းလမ်းပါ။ Field တစ်ခုချင်းစီကို သူက ဘာအကြောင်းအရာလဲဆိုတာ သိအောင် ခေါင်းစဉ် (Label) လေးတွေ တပ်ပြီး ပေါင်းစပ်လိုက်တာ ဖြစ်ပါတယ်။

သင့်ရဲ့ `usp_getInventoryByLocation` JSON ကို ဥပမာအနေနဲ့ ကြည့်ရအောင်။

**မူရင်း JSON:**
```json
{
  "ProcedureName": "usp_getInventoryByLocation",
  "Description": "ဆိုင်ခွဲတစ်ခု၏ လက်ရှိပစ္စည်းစာရင်းကို ရှာဖွေပေးသည်။ ကုန်ပစ္စည်းလက်ကျန်၊ stock level နှင့် ပစ္စည်းအရေအတွက်ကို သိရှိရန်သုံးသည်။",
  "Keywords": ["inventory", "stock", "ပစ္စည်းလက်ကျန်", "ကုန်ကျန်", "ဆိုင်ခွဲ"]
}
```

**Embedding အတွက် ပြင်ဆင်ရမည့် စာသားပုံစံ:**
```text
Procedure: usp_getInventoryByLocation.
Description: ဆိုင်ခွဲတစ်ခု၏ လက်ရှိပစ္စည်းစာရင်းကို ရှာဖွေပေးသည်။ ကုန်ပစ္စည်းလက်ကျန်၊ stock level နှင့် ပစ္စည်းအရေအတွက်ကို သိရှိရန်သုံးသည်။
Keywords: inventory, stock, ပစ္စည်းလက်ကျန်, ကုန်ကျန်, ဆိုင်ခွဲ.
```

**ဘာကြောင့် ပိုကောင်းတာလဲ။**
*   **Context (အကြောင်းအရာ) ပိုပြည့်စုံတယ်:** `usp_getInventoryByLocation` ဆိုတာ Procedure ရဲ့ နာမည်ဖြစ်ကြောင်း၊ "ဆိုင်ခွဲတစ်ခု..." ဆိုတာက ဖော်ပြချက်ဖြစ်ကြောင်း Model က ခွဲခြားသိမြင်နိုင်သွားတယ်။
*   **အဓိပ္ပာယ် ရောထွေးမှုမရှိဘူး:** စာသားအားလုံးကို ဒီအတိုင်း ရောချလိုက်တာထက် ပိုပြီး စနစ်တကျရှိတာကြောင့် Vector က ပိုတိကျပါတယ်။

### Keyword များကို ပိုထိရောက်အောင် ရေးသားနည်း

Format မှန်သွားပြီဆိုရင် အထဲမှာ ထည့်သွင်းမယ့် Keyword တွေက အရေးကြီးလာပါတယ်။

#### ၁။ နည်းပညာစကားလုံး (Technical) နှင့် သုံးနေကျစကားလုံး (Natural) နှစ်မျိုးလုံးထည့်ပါ။
*   **ဥပမာ:** `usp_getInventoryByLocation` အတွက်
    *   **Technical:** `inventory`, `stock level`
    *   **Natural:** `ပစ္စည်းဘယ်လောက်ကျန်လဲ`, `ကုန်စာရင်း`, `လက်ကျန်`

#### ၂။ စကားလုံးတစ်လုံးချင်းထက် စကားစု (Phrase) များကို သုံးပါ။
User တွေက စကားလုံးတစ်လုံးတည်း ရိုက်ရှာတာထက် စကားစုနဲ့ ရှာတာ ပိုများပါတယ်။
*   **မကောင်းသော Keyword:** `supplier`, `အသစ်`, `ထည့်`
*   **ကောင်းသော Keyword:** `supplier အသစ်ထည့်`, `ကုန်သွင်းသူ စာရင်းသွင်း`

#### ၃။ User တစ်ယောက်လို စဉ်းစားပြီး မေးနိုင်မယ့် မေးခွန်းများ (Hypothetical Questions) ထည့်ပါ။
ဒါက အလွန်အစွမ်းထက်တဲ့ နည်းလမ်းပါ။ သင့် JSON မှာ `SampleQueries` ဆိုတဲ့ field အသစ်တစ်ခု ထပ်ထည့်ပါ။
```json
{
  "ProcedureName": "usp_addNewSupplier",
  "Description": "ကုန်ပစ္စည်းအသစ်သွင်းရန်အတွက် supplier အသစ်တစ်ဦး၏ အချက်အလက်များကို မှတ်တမ်းတင်သည်။",
  "Keywords": ["supplier", "ကုန်သွင်းသူ", "ပစ္စည်းပေးသူ", "အသစ်ထည့်", "supplier စာရင်းသွင်း"],
  "SampleQueries": [
    "Supplier အသစ် ဘယ်လိုထည့်ရမလဲ။",
    "ပစ္စည်းပေးသွင်းသူအသစ် မှတ်တမ်းတင်ချင်တယ်။",
    "ကုန်သွင်းသူ list ထဲကို အသစ်ထပ်ထည့်လို့ရလား။"
  ]
}
```
ပြီးရင် အဲဒီ `SampleQueries` တွေကိုပါ ပေါင်းပြီး Embedding လုပ်ရမယ့် စာသားတုံးထဲ ထည့်လိုက်ပါ။

**ပြင်ဆင်ပြီး Embedding လုပ်ရမည့် စာသားပုံစံ အပြည့်အစုံ:**
```text
Procedure: usp_addNewSupplier.
Description: ကုန်ပစ္စည်းအသစ်သွင်းရန်အတွက် supplier အသစ်တစ်ဦး၏ အချက်အလက်များကို မှတ်တမ်းတင်သည်။
Keywords: supplier, ကုန်သွင်းသူ, ပစ္စည်းပေးသူ, အသစ်ထည့်, supplier စာရင်းသွင်း.
Sample Queries: Supplier အသစ် ဘယ်လိုထည့်ရမလဲ။ ပစ္စည်းပေးသွင်းသူအသစ် မှတ်တမ်းတင်ချင်တယ်။ ကုန်သွင်းသူ list ထဲကို အသစ်ထပ်ထည့်လို့ရလား။
```

### အနှစ်ချုပ်: သင့်လျော်သောနည်းလမ်းများ နှင့် ရှောင်ရန်အချက်များ

| သင့်လျော်သောနည်းလမ်းများ (Do) | ရှောင်ရန်အချက်များ (Don't) |
| :--- | :--- |
| ✅ **Label တပ်ပြီး ပေါင်းစပ်ပါ။** (`Procedure:`, `Description:`) | ❌ JSON field တွေကို ဒီအတိုင်း ရောချပြီး မထည့်ပါနဲ့။ |
| ✅ **User မေးနိုင်မယ့် မေးခွန်းအပြည့်အစုံ** ကို `SampleQueries` အဖြစ်ထည့်ပါ။ | ❌ Keyword တွေကို စကားလုံး တစ်လုံးချင်းစီ ခွဲမထည့်ပါနဲ့။ (`အရောင်း`, `သွက်` အစား `အရောင်းသွက်` လို့ထည့်ပါ) |
| ✅ **နည်းပညာစကားလုံးရော၊ သုံးနေကျစကားလုံးပါ** ထည့်ပါ။ (`stock` ရော `လက်ကျန်` ပါ) | ❌ Description ကို အရမ်းရှည်လျားပြီး မလိုအပ်တဲ့ စာတွေနဲ့ မရေးပါနဲ့။ တိုပြီး လိုရင်းကိုရေးပါ။ |
| ✅ **Field တစ်ခုနဲ့တစ်ခုကို Newline (`\n`)** သို့မဟုတ် **Period & Space (`. `)** နဲ့ ခြားပေးပါ။ | ❌ Embedding လုပ်မယ့် စာသားထဲမှာ JSON ရဲ့ `{}` `[]` `"` သင်္ကေတတွေ မပါအောင် သတိထားပါ။ |

ဒီနည်းလမ်းတွေကို သုံးပြီး သင့်ရဲ့ JSON file တွေကို ပြန်လည်ပြင်ဆင်၊ ပြီးမှ Embedding အသစ်ပြန်လုပ်ပြီး စမ်းသပ်ကြည့်ပါ။ သင့် RAG system ရဲ့ တိကျမှု သိသိသာသာ တက်လာတာကို တွေ့မြင်ရပါလိမ့်မယ်။
